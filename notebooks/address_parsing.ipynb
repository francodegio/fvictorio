```python
import pandas as pd
```

```python
df = pd.read_csv('test2.csv')
```

```python
mapper = {
    'CALLE': (0, 31),
    'CODIGO POSTAL': (31,35),
    'LOCALIDAD': (37,56),
    'PROVINCIA':  (56, -1)
}
```

```python
for name, (start, finish) in mapper.items():
    df[name] = df['DOMICILIO'].apply(lambda x: x[start:finish])
```
